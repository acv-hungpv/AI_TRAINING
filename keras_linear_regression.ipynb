```python
import numpy as np
np.random.seed(1000)
```

```python
from keras.models import Sequential
from keras.layers import Dense 
import matplotlib.pyplot as plt
from keras import optimizers
```

```python
X_train = np.asarray([3.3, 4.4, 5.5, 6.71, 6.93, 4.168, 9.779, 6.182, 7.59, 2.167,
                      7.042, 10.791, 5.313, 7.997, 5.654, 9.27, 3.1])
Y_train = np.asarray([1.7, 2.76, 2.09, 3.19, 1.694, 1.573, 3.366, 2.596, 2.53, 1.221,
                      2.827, 3.465, 1.65, 2.904, 2.42, 2.94, 1.3])
n_samples = X_train.shape[0]
print(n_samples)
plt.plot(X_train, Y_train, 'ro')
```

```python
model = Sequential()
model.add(Dense(output_dim=1, input_dim=1, activation='linear'))
model.summary()
```

```python
sgd = optimizers.SGD(lr=0.01)
model.compile(loss='mse', optimizer=sgd)
model.fit(X_train,Y_train,epochs=500)
```

```python
y_pred = model.predict(X_train)
plt.plot(X_train, Y_train, 'ro', label='Original data')
plt.plot(X_train, y_pred, label='Fitted line')
plt.legend()
plt.show()   
```
